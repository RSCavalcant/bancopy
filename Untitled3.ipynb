```python
df = pd.read_excel("AdventureWorks.xlsx")
```

```python
df.head()
```

```python
df["Valor Venda"].sum()
```

```python
df["Custo"] = df["Custo Unitário"].mul(df["Quantidade"])
```

```python
df.head(1)
```

```python
df["Custo"].sum()
```

```python
df["Lucrot"] = df["Valor Venda"] - df["Custo"]
```

```python
df["Lucrot"].sum()
```

```python
df["Tempo_envio"] = (df["Data Envio"] - df["Data Venda"]).dt.days
```

```python
#Média do tempo de envio por Marca
df.groupby("Marca")["Tempo_envio"].mean()
```

```python
produtos_mais_desconto = df.loc[df.groupby("Marca")["Valor Desconto"].idxmax()]
colunas_desejadas = ["ID Loja", "ID Produto","Produto", "Valor Venda", "Valor Desconto", "Fabricante"]  # Lista com os nomes das colunas desejadas
print(produtos_mais_desconto[colunas_desejadas])
```

```python
menor_venda=df.loc[df.groupby("Marca")["Quantidade"].idxmin()]
colunas_desejadas = ["ID Loja", "ID Produto","Produto", "Valor Venda", "Valor Desconto", "Fabricante", "Quantidade"]  # Lista com os nomes das colunas desejadas
print(menor_venda[colunas_desejadas])
```
